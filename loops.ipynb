```python
# Write a number guessing game where:
# The program generates a random number between 1 and 100
# The user has to guess the number
# After each guess, the program tells the user if their guess is too high or too low
# The game continues until the user guesses correctly
# Display the number of attempts taken
import random 
num=random.randint(1, 100)
attempts=0
while True:
    guess=int(input("enter your guess between 1-100:"))
    if guess>num:
        print("too high")
        attempts+=1
    elif guess<num:
        print("too low")
        attempts+=1
    else:
        num==guess
        print("found")
        attempts+=1
        break
print("attempts taken:",attempts)
print("numbers was:",num)
    

```

```python
# Write a program to check if a given number is a palindrome or not.
num=int(input("enter the number:"))
reverse=0
while num>0:
    digit=num%10
    reverse=reverse*10+digit
    num=num//10
if num==reverse:
    print("it is a palindrome")
else:
    print("it is not a palindrome")
```

```python
# Write a program to check if a number is an Armstrong number (also called Narcissistic number).
num=int(input("enter the number:"))
original=num
arms=0
n=len(str(num))
while num>0:
    digit=num%10
    arms=arms+(digit**n)
    num=num//10
if arms==original:
    print("it is armstrong")
else:
    print("it is not armstrong")
```
